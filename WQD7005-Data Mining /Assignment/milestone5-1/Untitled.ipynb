```python
# data crawling

import requests
import pandas as pd
from bs4 import BeautifulSoup

headers={
    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'
}


url = "https://www.indexmundi.com/commodities/?commodity=petroleum-price-index&months=360"

res = requests.get(url,headers=headers)

try:

    df = pd.read_html(res.text)
    pan = pd.DataFrame(df[1])
    data = {'Month':pan.iloc[:,0],'Price':pan.iloc[:,1],'Change':pan.iloc[:,2]}
    df = pd.DataFrame(data=data,columns=['Month', 'Price', 'Change'])
    df["Change"]=list(df["Change"].apply(lambda x:x[:-1]).values)
    df.to_csv('oil_price.csv')
    
except:
    print("The url doesn't work!")
```
