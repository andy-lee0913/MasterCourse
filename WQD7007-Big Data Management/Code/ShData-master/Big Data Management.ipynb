# Who is the best seller ?

```python
import plotly.express as px
import pandas as pd
df=pd.read_csv("samples/Automotive.csv")
```

```python
df.head(10)
```

```python
fig = px.bar(df, x="Seller", y="SellerRatings")
fig.show()
```

```python
fig = px.bar(df, x="Seller", y="Followers")
fig.show()
```

```python
fig = px.bar(df, x="Seller", y="ResponseRate")
fig.show()
```

```python
import plotly.graph_objs as go



trace1 = go.Bar(
   x = df.Seller,
   y = df.ResponseRate,
   name = 'FY'
)

data = [trace1]

layout = go.Layout(barmode = 'group')

fig = go.Figure(data = data, layout = layout)
fig.show()
```
